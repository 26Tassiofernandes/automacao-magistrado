### Iniciar o Selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

navegador.get(r"D:\\45577\Documents\\Coding\\automacao-magistrado\\projeto-magistrados\src\\index.html") # Aqui botamos o caminho relativo do arquivo
```

### Importar a tabela excel e preencher os campos desejados

```python
import pandas as pd
from time import sleep

tabela_magistrados = pd.read_csv(r'magistrados-pje.csv')
# display(tabela_magistrados)

contagem = 0

for linha in tabela_magistrados.index:
    
    # Preencher cpf:
    navegador.find_element(By.ID, "cpf").send_keys(tabela_magistrados.loc[linha, "cpf"])
    
    # Preencher email:
    navegador.find_element(By.ID, "email").send_keys(tabela_magistrados.loc[linha, "dispositivo_intimacao"])
    
    contagem += 1

    cpf = tabela_magistrados.loc[linha, "cpf"]
    email = tabela_magistrados.loc[linha, "dispositivo_intimacao"]
    matricula = tabela_magistrados.loc[linha, "matricula"]
    print("CPF: {}\nE-mail: {}\nMatrícula: {}\n\n".format(cpf, email, matricula))
    
    navegador.refresh()
print('Ao todo, foram', contagem , 'magistrados cadastrados!')
```
